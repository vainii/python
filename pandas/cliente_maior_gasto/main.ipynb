```python
import pandas as pd

clientes = pd.read_csv("clientes.csv")
vendas = pd.read_csv("vendas.csv")
```

```python
clientes_vendas = pd.merge(clientes, vendas, on = "cliente_id")
clientes_vendas.head()
```

```python
clientes_gastos = clientes_vendas.groupby("nome")["valor"].sum().reset_index().sort_values("valor", ascending = False)
clientes_gastos.head()
```
