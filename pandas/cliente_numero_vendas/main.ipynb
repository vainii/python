```python
import pandas as pd 

clientes = pd.read_csv("clientes.csv")
vendas = pd.read_csv("vendas.csv")

```

```python
vendas[vendas.duplicated(keep=False)]
```

```python
clientes_vendas = pd.merge(clientes, vendas, on = "cliente_id")

clientes_vendas.drop_duplicates(subset = ["venda_id"], inplace = True)
```

```python
clientes_total_vendas = clientes_vendas.groupby("nome")["venda_id"].count().reset_index().sort_values("venda_id", ascending = False)
clientes_total_vendas.rename(columns = {"venda_id" : "total_vendas"}, inplace = True)
print(clientes_total_vendas)
```
