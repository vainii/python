```python
import pandas as pd

clientes = pd.read_csv('clientes.csv')
vendas = pd.read_csv('vendas.csv')
```

```python
clientes_vendas = pd.merge(clientes, vendas, on='cliente_id')
clientes_vendas.head()
```

```python
clientes_vendas.rename(columns={'cidade' : 'estado'}, inplace=True)

clientes_vendas['valor_total'] = clientes_vendas['quantidade'] * clientes_vendas['valor']

clientes_vendas.groupby("estado", as_index = False)["valor_total"].sum().sort_values("valor_total", ascending = False)
```
