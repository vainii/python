```python
import pandas as pd 

clientes = pd.read_csv('clientes.csv')
vendas = pd.read_csv('vendas.csv')
```

```python
clientes_vendas = pd.merge(clientes, vendas, on = 'cliente_id')
clientes_vendas[clientes_vendas.duplicated(subset = 'venda_id', keep = False)].sort_values('venda_id')
```

```python
clientes_vendas.drop_duplicates(subset = 'venda_id', inplace = True)
clientes_vendas.sort_values('venda_id')
```

```python
clientes_vendas.to_csv('clientes_vendas.csv', index = False)
```
